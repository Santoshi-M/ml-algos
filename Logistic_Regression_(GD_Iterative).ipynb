```python
import sklearn
import math
```

```python
X, Y_np = sklearn.datasets.load_breast_cancer(return_X_y = True)
```

```python
m = len(X)
print("m: ", m)
```

```python
n = len(X[0])
print("n: ", n)
```

```python
# Z = WX + B
# Yhat = sig(Z)

# log loss
# L = - (Y * log(Yhat) +  (1-Y) * log(1-Yhat))

# while L > acceptable_loss:
  # dL/dZ = dZ = Yhat - Y
  # dW = (1/m) * (X * dZ)
  # dB = (1/m) * dZ

  # W = W - learning_rate * dW
  # B = B = learning_rate * dB
```

```python
# initialise

W = [0.0] * n
B = 0.0
Z = [float(0)] * m
Yhat = [float(0)] * m
learning_rate = 0.0001
acceptable_loss = -1 * math.exp(6)
L = 1000

dZ = [float(0)] * m
dW = [float(0)] * m
dB = 0.0
iter = 0

Y = [int(ele) for ele in Y_np]
```

```python
# Gradient Descent

while iter < 1000:
  iter += 1
  L = 0
  Z = [float(0)] * m

  for i in range(m):
    # compute preds
    for j in range(n):
      Z[i] += (W[j] * float(X[i][j])) + B

    try:
      Yhat[i] = 1.0 / (1.0 + math.exp(-Z[i]))
    except OverflowError:
      Yhat[i] = 0

    # compute loss
    if Y[i] == 1.0:
      if Yhat[i] == 0:
        L += 1000
      else:
        L += (-1) * math.log(Yhat[i])
    else:
      if Yhat[i] == 1:
        L += 1000
      else:
        L += (-1) * math.log(1 - Yhat[i])

    # compute gradients
    dZ[i] = Yhat[i] - Y[i]
    dB += dZ[i]
    for j in range(n):
      dW[j] += (X[i][j] * dZ[i])

  L /= m
  for j in range(n):
      dW[j] /= m
  dB /= m
  print(f"iter: {iter} L: {L}")

  # update weights
  for j in range(n):
    W[j] = W[j] - (learning_rate * dW[j])
    B = B - (learning_rate * dB)

```

```python
pred = [float(0)] * m
label = [float(0)] * m
accuracy = 0

for i in range(m):
  pred[i] = 0
  for j in range(n):
    pred[i] += (X[i][j] *  W[j])
  pred[i] += B
  try:
    pred[i] = 1.0 / (1 + math.exp(-pred[i]))
  except OverflowError:
    pred[i] = 0.0

  label[i] = Y[i]

  pred[i] = 1.0 if pred[i] > 0.5 else 0.0

  if pred[i] == label[i]:
    accuracy += 1
accuracy /= m
accuracy *= 100

```

```python
print(f"accuracy: {round(accuracy, 2)}%")
```
