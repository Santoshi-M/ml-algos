```python
import sklearn
import math
import numpy as np
```

```python
X, Y = sklearn.datasets.load_breast_cancer(return_X_y = True)
```

```python
m = len(X)
n = len(X[0])
print(f"m: {m}, n: {n}")
```

```python
Y = Y.reshape(569,1)
```

```python
# Z = WX + B
# Yhat = sig(Z)

# log loss
# L = - (Y * log(Yhat) +  (1-Y) * log(1-Yhat))

# while num_iter < K:
  # dL/dZ = dZ = Yhat - Y
  # dW = (1/m) * (X * dZ)
  # dB = (1/m) * dZ

  # W = W - learning_rate * dW
  # B = B = learning_rate * dB
```

```python
# initialise

W = np.zeros(n).reshape(1, n)
B = 0.0
learning_rate = 0.0001
L = 1000
iter = 0

def stable_sigmoid(x):
    if x >= 0:
        return 1.0 / (1 + np.exp(-x))
    else:
        return np.exp(x) / (1 + np.exp(x))
```

```python
# Gradient Descent

while iter < 1000:
  iter += 1
  L = 0

  Z = np.dot(X, np.transpose(W)) + B
  Yhat = np.array([stable_sigmoid(elem) for elem in Z])

  # compute loss
  for i in range(m):
    if Y[i] == 1.0:
      if Yhat[i] == 0:
        L += 1000
      else:
        L += (-1) * math.log(Yhat[i])
    else:
      if Yhat[i] == 1:
        L += 1000
      else:
        L += (-1) * math.log(1 - Yhat[i])

  # compute gradients
  dZ = Yhat - Y
  dB = dZ
  dW = np.dot(np.transpose(dZ), X)

  L /= m
  dW /= m
  dB /= m
  print(f"iter: {iter} L: {L}")

  # update weights
  W = W - learning_rate * dW
  B = B - learning_rate * dB

```

```python
pred = np.dot(X, np.transpose(W)) + B
pred = np.array([stable_sigmoid(elem) for elem in Z])
label = Y
pred = np.where(pred > 0.5, 1.0, 0.0)
accuracy = np.average(pred == label)
accuracy *= 100
```

```python
print(f"accuracy: {round(accuracy, 2)}%")
```
